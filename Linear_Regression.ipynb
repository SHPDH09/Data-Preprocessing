```python
# prompt: remove outlire

import numpy as np

Q1 = df[['newspaper']].quantile(0.25)
Q3 = df[['newspaper']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[~((df[['newspaper']] < lower_bound) | (df[['newspaper']] > upper_bound)).any(axis=1)]

# Print the DataFrame without outliers
print(df_no_outliers)

plt.figure(figsize=(8, 6))
df_no_outliers[['TV', 'radio', 'newspaper']].boxplot()
plt.title('Boxplot of Advertising Channels (Without Outliers)')
plt.ylabel('Expenditure')
plt.show()
```
